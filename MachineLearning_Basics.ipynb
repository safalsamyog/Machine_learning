**Z Score**
1. it is method for removing the outliers present in the data

```python
mean1=data_set2.Area.mean()
mean1#showing very unacurate mean
```

```python
std1=data_set2.Area.std()
std1
```

**IQR Outlier Remover**

g.png

```python
#IQR=Q3-Q1
#lower_limit=Q1-1.5*IQR
#upper_limit=Q3+1.5*IQR
#outliers is less than (lower_limit) or greater than (upper_limit)
```

```python
data_set=pd.read_csv('house.csv')
data_set.head()
```

```python
d2=data_set.fillna({'Area':data_set.Area.median(),'Price':data_set.Price.median()})#this dicitonary method to fill the data we can use mean , repating value or median 
d2
```

```python
d2.describe()
```

```python
Q1=d2.Area.quantile(0.25)
Q3=d2.Area.quantile(0.75)
Q1,Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
lower_limit,upper_limit
```

```python
d2[(d2.Area<lower_limit)|(d2.Area>upper_limit)]#we easily detected outliers more acuurate than others 
```

```python
non_outliers=d2[(d2.Area>lower_limit)&(d2.Area<upper_limit)]#use and sign becuase we need to satisfy both condition
non_outliers
```

# **CDF and Pdf**

```python
#cumelative disribution function and Probality distribution function
```

```python
count,bin=np.histogram(set['petal.length'],bins=10)#bins means difference which genrate us 10 values
print(count,bin)
```

```python
pdf=count/(sum(count))
pdf
```

```python
cdf=np.cumsum(pdf)
cdf
```

```python
plt.figure(figsize=(9,5))
plt.title('Setosa Petal Length')
plt.plot(bin[1:],pdf)
plt.plot(bin[1:],cdf)
plt.xlabel('petal length of setosa')
plt.ylabel('Probality')
```

the petal length having 1.5 is 60% probality of setosa and having 1.9 100% probality
