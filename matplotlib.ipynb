```python
import numpy as np
import matplotlib.pyplot as plt

x=[]
for i in range(1,8):
    xx=int(input("enter number: "))
    x.append(xx)


y=[]
for j in x:
    yy=j**2-6*j+8#parabola equation y=x**2-6x+8
    y.append(yy)

plt.plot(np.array(x),np.array(y),'b')

plt.title("parabola")
plt.xlabel("X-Axis",{"size":20,"color":"r"})
plt.ylabel("Y-Axis",{"size":20,"color":"r"})
plt.show()    
# plt.plot(x,y)
# plt.show()
    
    
```

```python

x=[0,30,60,90,120,150,180,210,240,270,330,360,390,420,450,480,510,540,570,630,660,690,720,750,780,810,840,870,930,960,990,1020,1050,1080,1150,1180,1210,1240,1270,1330]

y=[]
yy=[]
# for i in range(1,20+4):
#     xx=int(input("Enter the degrees: "))
#     x.append(xx)

for z in x:
    k=np.sin(np.around(z,decimals=2)*np.pi/180)
    kk=np.cos(np.around(z,decimals=2)*np.pi/180)
    y.append(k)
    yy.append(kk)
# x=np.array([0,30,60,90,120,150,180,210,240,270,330,360])
# y=np.array([0.00,0.5,0.87,1.00,0.87,0.5,0.00,-0.5,-0.87,-1.0,-0.5,0.00])
plt.plot(x,y,"r",label="sin",linewidth=1.5,zorder=2,linestyle="--",marker=".")
plt.plot(x,yy,"g",label="cos",linewidth=1.5,zorder=1,alpha=0.79,marker=".",linestyle="--")
plt.grid(b=True)
plt.title("cosine and sine")
plt.show()
```

```python
#stack plot
x=np.array([2,3,4,6])
y=np.array([5,7,9,8])
z=np.array([6,5,8,9])
a=np.array([9,10,16,18])
plt.title("Area PLot")
plt.plot([],[],color="r",linewidth=5,label="eating")
plt.plot([],[],color="b",linewidth=5,label="sleping")
plt.plot([],[],color="g",linewidth=5,label="watching")
plt.plot([],[],color="c",linewidth=5,label="learning")
plt.stackplot(x,y,z,a,colors=["r","b","g","c"])
plt.legend()
plt.show()
```
