```python
import pandas as pd
dataset=pd.read_csv('data1.csv')
data_set=dataset.drop(10)
data_set2=data_set.drop(columns='Age')
data_set2
x=data_set2.iloc[:,0].values
y=data_set2.iloc[:,-1].values
data_set2
```

```python

m=0
c=0
l=0.00001
epochs=1000

n=len(x)
for i in range(epochs):
    y_pred=m*x+c
    Dm=(-2/n)*sum(x*(y-y_pred))
    Dc=(-2/n)*sum(y-y_pred)
    m=m-(l*Dm)
    c=c-(l*Dc)
print(m,c)
```

```python
import matplotlib.pyplot as plt
x=data_set2.iloc[:,0].values
y=data_set2.iloc[:,-1].values

y_pred=m*x+c
plt.scatter(x,y,label='actual')
plt.plot(x,y_pred,label='pred',color='r')
plt.legend()
```

```python
data1=data[data.cluster==0]
data2=data[data.cluster==1]
data3=data[data.cluster==2]
```
